## 甚麼是 Flow control (流程控制)?

## 04.2 if / else / elif

## if 的功能與介紹
### if是一個能夠幫助我們更靈活實現流程控制的工具，我們可以用if設定條件，只要條件達成便會執行if集區(Block)下的程式碼，如以下範例所示，我們會以空格來區分集區。

```python
name = 'Alice'
if name=='Alice':
  print('Hello')
```

### 除此之外，我們還可以加入elif或else來讓整個流程控制的過程更有變化，當if後的條件沒有達成時會接著判斷是否符合elif的條件，而else則是當上述條件都沒有達成時就會執行else集區底下的程式碼讓我們的程式碼能夠根據不同情況應對更多不同的邏輯判斷條件。


```python
name = 'Alex'
if name=='Alice':
  print('Hello')
elif name=='Alex':
  print('Hi')
else:
  print('Done')
```

```python
name = 'Stephen'
if name=='Alice':
  print('Hello')
elif name=='Alex':
  print('Hi')
else:
  print('Done')
```

```python
name = 'Alex'
age =  3000
if name=='Alice':
  print('Hi Alice')
elif age<12:
  print('You are not Alice, kiddo.')
elif age>1000:
  print('Not like you, Alice is not an nudead vampire.')
elif age>100:
  print('You are not Alice, grannie.')
```

## 04.3 While 迴圈
### While迴圈能夠透過關係運算式控制迴圈是否執行。當該關係運算式的結果是True的時候，才會執行迴圈中的程式碼。如果是False，就會離開迴圈。下圖是While迴圈的運作流程圖。


```python
x = 0
while x<5:
  print('Hello!')
  x = x + 1
```

```python
name = ''
while name!='your name':
  print('Please input your name.')
  name = input()
print('Thank you!')
```

## 無限迴圈
### 有的時候我們會設定錯誤的條件讓我們不小心創造出了無限迴圈，這時我們就需要強制終止程式。如果是使用Python內建的IDLE則可以按Ctrl+C來強制終止程序。



```python
x = 3
while x==3:
  print('Hello!')
```

### 我們也可以在集區裡面加入break指令來強制終止迴圈

```python
x = 3
while x==3:
  print('Hello!')
  break
```

### 透過continue指令控制迴圈跳過某個執行階段

```python
x = 0
while x<5:
  x=x+1
  if x==3:
    continue
  print(x)
```

## 04.4 for 迴圈
### for迴圈是Python的第二種迴圈，我們同樣可以透過設定判斷條件來讓for迴圈重複執行流程，和while迴圈相同，我們可以在集區中加入continue或break來控制迴圈的執行程序。


## range()函數
### range是常常伴隨for迴圈一起使用的函數，這裡簡單的介紹一下，需要特別注意的是range()返回的是特殊形式的資料格式而非列表(list)，因此我們不能直接用print()打印出資料內的數字。

```python
x = range(0,10)
type(x)
```

```python
print(x)
```

```python
for i in range(0,5):
  print(i)
```

### 我們還可以進一部指定range()函數內的數字間隔，例如:



```python
for i in range(0,10,2):
  print(i)
```

```python
for i in range(5,-1,-1):
  print(i)
```

### for 迴圈範例

```python
for i in range(0,10):
  print(i)
```

```python
print('My name is : ')
for i in range(0,5):
  print('Alex Five Time '+str(i))
```

```python
total = 0
for number in range(1,101):
  total = total + number
print(total)
```

```python
for i in range(0,5,1):
  print(i)
```
